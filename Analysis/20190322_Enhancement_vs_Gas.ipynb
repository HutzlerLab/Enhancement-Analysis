```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
import numpy as np
import statistics as stat
from enhancement_analysis import *
from fit_functions import *
from low_level import readCSVrows,writeCSVrows,calcEnhancement,readCSVcolumns
%matplotlib inline
```

```python
results=readCSVcolumns('Processed_20190322/data_legend_sccm.csv',skip_header=1)
header = results[1]
print(header)
[sccm,starts,stops]= results[0]
```

```python
folder = '20190322'
skips = []
ABAB = True

start_arr = starts.astype(int)
stop_arr = stops.astype(int)
sccm_arr = sccm

enhancement = []

b_ODs = []
ub_ODs = []

for _start,_stop in zip(start_arr,stop_arr):
    
    ODs_1 = calculateSeriesOD(folder,_start,_stop,skips,ABAB=ABAB)
    ODs_2 = calculateSeriesOD(folder,_start+1,_stop,skips,ABAB=ABAB)
    
    p_1 = processData(ODs_1,[0,8])
    p_2 = processData(ODs_2,[0,8])
    b,ub = identifyBUB(p_1,p_2,equal=True)
    en = calcEnhancement(b[0],ub[0])
    enhancement.append(en)
    b_ODs.append(b)
    ub_ODs.append(ub)
#p=YbOH integrated, Yb integrated, timestamp
```

```python
avg = []
err = []
for _en,_s in zip(enhancement,sccm):
    plt.figure()
    plt.plot(_en,label='{} SCCM'.format(_s))
    plt.legend(loc='best')
    avg.append(np.array(_en).mean())
    err.append(np.array(_en).std())
```

```python
en_sort = np.array(sortData(sccm,avg))
enerr_sort = np.array(sortData(sccm,err))
sccm_sort = np.array(sorted(sccm))
```

```python
data = np.array([np.array(sccm_sort),np.array(en_sort),np.array(enerr_sort)])
data.T
```

```python
lbls = ['Gas Flow (SCCM)','Enhancement (OD Ratio)','Sigma']
data = np.array([np.array(sccm_sort),np.array(en_sort),np.array(enerr_sort)])
save_all = []
save_all.append(lbls)
for _data in data.T:
    save_all.append(_data)
```

```python
writeCSVrows(save_all,'Processed_20190322/avg_enhancement.csv')
```

```python
plt.figure()
plt.title('Enhancement vs Flow')
plt.xlabel('Gas Flow (SCCM)')
plt.ylabel('OD Ratio')
plt.errorbar(sccm_sort,en_sort,yerr=enerr_sort,marker='o',linestyle='None')
```
